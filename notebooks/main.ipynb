```python
import pandas as pd

from mailsift.config import DATASETS_DIR
```

```python
sources = ["CEAS_08", "Enron", "Ling", "SpamAssasin"]
frames = []
for source in sources:
    df = pd.read_csv(f"{DATASETS_DIR}/dataset-1/{source}.csv",
                     usecols=["subject", "body", "label"],
                     dtype={"subject": str, "body": str, "label": int},
                    #  nrows=2859
                     )
    print(f"{source}: {len(df)} emails")
    frames.append(df)

dataset = pd.concat(frames, ignore_index=True)
print(f"Total: {len(dataset)} emails")



    
```

```python
dataset["body"] = dataset["body"].fillna("")

dataset["body"] = dataset["body"].apply(lambda x : x.replace('\n', ' ').replace('\r', ' '))
dataset
```
